```python
#In this Python file I imported a Red Wine dataset and try to answered a few random question about the dataset by plotting 
#the data

#Importing libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Importing the Red Wine dataset
df_wine = pd.read_csv('Red.csv')

#Reading dataset
df_wine
```

```python
#Question 1: Which country has the highest number of user ratings?

#Plotting number of ratings for each country
plt.figure(figsize=(18,10))
plt.title('Number of Rating per Country')
plt.barh('Country', 'NumberOfRatings', data = df_wine, color = 'darkorange')
plt.xlabel('Country')
plt.ylabel('Number Of Rating')
```

```python
#Question 3: What is the top 10 highest rated wines in Spain?

#Creating new dataset that contains all of the wine that was rated in Spain
df_spain = df_wine.loc[df_wine['Country'] == 'Spain']
df_spain_rate = df_spain.sort_values('Rating', ascending=False).head(10)
df_spain_rate
```

```python
#Plotting a Horizontal Bar Chart for top 10 highest rated wine in Spain
x = df_spain_rate['Name']
y = df_spain_rate['Rating']

fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(x,y, 0.7, color='red')

for bar in bars:
  width = bar.get_width()
  label_y_pos = bar.get_y() + bar.get_height() / 2
  ax.text(width, label_y_pos, s=f'{width}', va='center')
for tick in ax.yaxis.get_major_ticks():
  tick.label.set_fontsize(10)

plt.title('Rating per Wine')
plt.xlabel('Rating')
plt.ylabel('Wine')
```

```python
#What are the top 5 most expensive wines?

#Creating smaller dataset with top 5 most expensive wine
df_price = df_wine.sort_values('Price', ascending=False).head(5)

#Plotting a Horizontal Bar Chart with the top 5 most expensive wine
x2 = df_price['Name']
y2 = df_price['Price']

fig, ax = plt.subplots(figsize=(15, 5))
bars = ax.barh(x2,y2, 0.7, color='Orange')

for bar in bars:
  width = bar.get_width()
  label_y_pos = bar.get_y() + bar.get_height() / 2
  ax.text(width, label_y_pos, s=f'{width}', va='center')
for tick in ax.yaxis.get_major_ticks():
  tick.label.set_fontsize(10)

plt.title('Top 5 Most expensive Wine')
plt.xlabel('Price (EUR)')
plt.ylabel('Wine')
```

```python
#Which year had the best User Rating for the Les Terrasses Velles Vinyes Priorat Red Wine?

#Plotting a Line Chart with the rating per year for the spesific wine
df_year = df_spain.loc[df_spain['Name'].str.contains('Les Terrasses Velles Vinyes Priorat')]
plt.figure(figsize=(15,8))
plt.plot('Year', 'Rating', data = df_year, color = 'red')
plt.title('Best Year for the Les Terrasses Velles Vinyes Priorat Wine')
plt.xlabel('Year')
plt.ylabel('Rating')
```
